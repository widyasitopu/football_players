8. The Player field of data frame is name of the football player, in which many of them are using Latin characters. However, there are unrecognized Latin characters which are replaced by a character â€˜?â€™. Define a function to get those players with character â€˜?â€™ in their names and save them into a dictionary containing {loc, name}.

```python
def players_with_question(new_data):
    players_with_question = []

    for index, row in new_data.iterrows():
        player_name = row['Player']

        if '?' in player_name:
            players_with_question.append({'loc': index, 'name': player_name})

    return players_with_question

players_with_question_dict = players_with_question(new_data)

for player_info in players_with_question_dict:
    print(f"Loc: {player_info['loc']}, Name: {player_info['name']}")
```

9. Print the dictionary and count how many players are there

```python
#print the dictionary
print(players_with_question_dict)
print("")

#the total of players
player_count = len(players_with_question_dict)
print(f"The total players with Question (?) Mark: {player_count}")
```

11. From collection import defaultdict library and define a function to map player based on player age.

```python
#import library defaultdict
from collections import defaultdict

#function map player based on player age
def map_players_by_age(new_data):
    players_by_age = defaultdict(list)

    for index, row in new_data.iterrows():
        player_name = row['Player']
        player_age = row['Age']

        players_by_age[player_age].append(player_name)

    return players_by_age

players_by_age_dict = map_players_by_age(new_data)

#sorting ascending
sorted_players_by_age = dict(sorted(players_by_age_dict.items()))

for age, players in sorted_players_by_age.items():
    print(f"Age: {age}, Player: {', '.join(players)}")
```

12. Get a list of certain playerâ€™s age and then print dictionary of playerâ€™s Nation: Squad

```python
#dictionary based on player age
def get_players_by_age(new_data, target_age):
    players_by_age_dict = {}

    for index, row in new_data.iterrows():
        player_name = row['Player']
        player_age = row['Age']
        player_nation = row['Nation']
        player_squad = row['Squad']

        if player_age == target_age:
            players_by_age_dict[player_name] = {'Nation': player_nation, 'Squad': player_squad}

    return players_by_age_dict

# get player based on a certain players age, example centain player's age = 25
target_age = 25
players_by_age_dict = get_players_by_age(new_data, target_age)

#print(players_by_age_dict)
for player_name, player_info in players_by_age_dict.items():
    print(f"Player Name: {player_name}")
    print(f"Nation: {player_info['Nation']}")
    print(f"Squad: {player_info['Squad']}")
    print()
```
